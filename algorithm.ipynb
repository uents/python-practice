### 1. リスト内の最大値を見つける
リスト内の最大値を見つけるには、

1. リストの先頭の値を `max_value` として覚える
2. リストから要素を取り出し、 `max_value` より大きいかをチェックする。大きければそれを `max_value` として覚え直す
3. 2.の処理を最後まで繰り返すと、リスト内の最大値を見つけることができる

````python
numbers = [1, 6, 3, 2, 4, 5]
max_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num

print('最大値:', max_value)
````

### 2. リスト内の表の数を数える
リスト内の `'表'` の数を数えるには、

1. 数え上げた数 `count` を0とする
2. リストから要素を取り出し、 `'表'` なら `count` を1つ増やす
3. 2.の処理を最後まで繰り返すと、リスト内の `'表'` の数を数えることができる

````python
sides = ['表', '裏', '裏', '表', '裏']
count = 0

for side in sides:
    if side == '表':
        count = count + 1

print('表の数:', count)
````

### 3. リスト内の奇数を探す
リスト内の奇数を探すには

1. 奇数を覚えさせる空のリスト `odds` を作る
2. リストから要素を取り出し、奇数なら `odds` に追加する
3. 2.の処理を最後まで繰り返すと、奇数のみのリストを作ることができる

````python
numbers = [1, 6, 3, 2, 4, 5]
odds = []  # 奇数のみのリスト

for num in numbers:
    if num % 2 == 1:
        odds.append(num)

print('奇数:', odds)
````

### 4. リストの合計値を求める
リストの合計値を求めるには、

* 合計値 `total` を、0とする
* リストから要素を取り出し、 `total` に足し合わせる
* 2.の処理を最後まで繰り返すと、リストの要素の合計を求めることができる

````python
numbers = [1, 6, 3, 2, 4, 5]
total = 0

for num in numbers:
    total = total + num

print('合計値:', total)
````

さらに、合計値を個数で割ると平均値を求めることができます。

````python
print('平均値:', total / len(numbers))
````

### 5. 西暦から元号を出力する

* 昭和元年は1926年
* 平成元年は1989年
* 令和元年は2019年

なので、 `if` 文を使うことで、西暦から元号を求めるプログラムを作成できます。

````python
year = 2012

if 1926 <= year < 1989:
    era = '昭和'
    era_year = year - 1925
elif 1989 <= year < 2019:
    era = '平成'
    era_year = year - 1988
elif year >= 2019:
    era = '令和'
    era_year = year - 2018

print(f'{year}年は、{era}{era_year}年です')
````

さらにこれを関数にしてみます。

````python
def gengou(year):
    if 1926 <= year < 1989:
        era = '昭和'
        era_year = year - 1925
    elif 1989 <= year < 2019:
        era = '平成'
        era_year = year - 1988
    elif year >= 2019:
        era = '令和'
        era_year = year - 2018

    # 元号と年をタプルにして返す
    return (era, f'{era_year}年')
````

### 6. 移動時間の計算
あなたは灰原哀です。「天国へのカウントダウン」で、火災中のA塔から隣塔のBへスーパーカーで飛び移るには、スーパーカーの速度がどのくらい必要か、いますぐに計算しなければいけません。

* 隣のビルの横方向の距離は60mです
* 隣のビルとの高さの差、つまりスーパーカーの落下距離は20mです



高さ方向の移動時間は、地球の重力によって決まります。



落下時間は下のプログラムで求めることができます。

````python
import math

height = 20  # A塔からB塔の高さの差
gravity = 9.80665  # 重力加速度

time = math.sqrt(2 * height / gravity)
print('落下時間 [秒]：', time)
````

落下時間の2秒で、隣のB塔に飛び移ればいいことがわかりました。

次に、2秒間で60m進む必要があるので、スーパーカーは秒速 30m が必要になります。

````python
length = 60  # A塔からB塔の距離 [m]

speed = length / int(time)  # スーパーカーに必要な速度（秒速）

print('速度 [秒速(メートル)]：', speed)
````

さらに、時速に直すには、1時間に何メートルを進むかを求めればよいので、

* 1分間だと、 30m * 60 = 1800m
* 1時間だと、 1800m * 60 = 108000m

これをkmに直すと、スーパーカーが必要な時速は、108km となります。

````python
print('速度 [時速(キロメートル)]：', speed * 60 * 60 / 1000)  # スーパーカーに必要な速度（時速）
````

#### 問題 6.1
落下する距離を引数とし、その落下時間を返す関数 `fall_time()` を作ります。

````python
import math

def fall_time(height):
    # 重力加速度
    gravity = 9.80665

    # 落下時間
    second = math.sqrt(2 * height / gravity)

    # 小数点3位で四捨五入する
    return round(second, 3)
````

### 問題 6.2
距離と速度と時間の関係は以下の通りです。

* 距離 = 速度 × 時間
* 時間 = 距離 / 速度
* 速度 = 距離 / 時間

つまり、速度は距離を時間で割ることで、時間は距離を速度で割ることで、求まります。

````python
# 速度と時間から、距離を求める
def length(speed, elapsed_time):
    return speed * elapsed_time

# 距離と速度から、時間を求める
def elapsed_time(length, speed):
    return length / speed

# 距離と時間から、速度を求める
def speed(length, elapsed_time):
    return length / elapsed_time
````
